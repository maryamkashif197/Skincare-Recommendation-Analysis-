```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

product_df = pd.read_csv('product_df.csv')

product_df.head()
```

```python
product_df.shape
```

```python
product_df.info()
```

```python
product_df.describe()
```

```python
display(product_df["brand"].unique())
```

```python

display( product_df["category"].unique())
```

```python
product_df["category"] = product_df["category"].replace("ANti-Aging", "Anti-Aging")
product_df["category"] = product_df["category"].replace("BB & CC Creams", "BB & CC Cream")
```

```python

display( product_df["category"].unique())
```

```python
product_df = product_df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
product_df.shape
```

```python

product_df.insert(0, 'id', range(len(product_df)))
```

```python
product_df.isnull().sum()
```

```python
product_df.shape
```

```python
display(product_df['size'].unique())
```

```python
ml_to_oz = 0.033814
size_value = 30
size_oz = round(size_value * ml_to_oz, 2)

size_oz_str = str(size_oz) + ' oz'

if (product_df['size'] == '30 mL').any():
    product_df['size'] = product_df['size'].replace('30 mL', size_oz_str)

print(size_oz_str)
```

```python
product_df['size'] = product_df['size'].str.replace(r'(\d+\.?\d*)oz', r'\1 oz', regex=True)
```

```python
product_df['size'] = product_df['size'].str.replace('•', '', regex=False)
```

```python
product_df['size'] = product_df['size'].fillna('Unknown')
```

```python
display(product_df['size'].unique())
```

```python
if product_df['category'].isnull().sum() > 0:
    product_df.dropna(subset=['category'], inplace=True)

product_df.shape
```

```python
# Find duplicate rows based on all columns except 'id'
duplicates = product_df[product_df.duplicated(subset=product_df.columns.difference(['id','n_of_loves']))]

if not duplicates.empty:
    print("Duplicate rows:")
    display(duplicates)
else:
    print("No duplicate rows found.")
```

```python
product_df.drop_duplicates(subset=product_df.columns.difference(['id','n_of_loves']), inplace=True)

product_df.shape
```

```python
display(product_df)
```

```python
print(product_df['price'].unique())
```

```python
product_df['price'] = product_df['price'].replace({
    '60.00(86.00 value)': '60.00',
    '30.00(45.00 value)': '30.00',
    '50.00(': '50.00'
})
```

```python

product_df['price'] = product_df['price'].astype(float)

```

```python
print(product_df['price'].unique())
```

```python
product_df = product_df.dropna(subset=['price'])
```

```python
display(product_df)
```
